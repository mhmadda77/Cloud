```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import ipywidgets as widgets
from IPython.display import display

students = []

with open('/content/drive/MyDrive/students.txt', 'r') as file:
    students = [line.strip() for line in file]

dropdown = widgets.Dropdown(options=students, description="Student:")

confirmButton = widgets.Button(description="Confirm")
editButton = widgets.Button(description="Edit")
nameInput = widgets.Text(description="Name:")
idInput = widgets.Text(description="ID:")

output = widgets.Output()

display(dropdown,confirmButton,nameInput,idInput,editButton,output)

def on_confirm(b):
    selected = dropdown.value
    name,id = selected.split(' ')

    nameInput.value = name
    idInput.value = id



def on_edit(b):
    updated_students = []

    with open('/content/drive/MyDrive/students.txt', 'r') as file:
        lines = [line.strip() for line in file]

    for line in lines:
        if line == dropdown.value:
            updated = f"{nameInput.value} {idInput.value}"
            updated_students.append(updated)
        else:
            updated_students.append(line)

    with open('/content/drive/MyDrive/students.txt', 'w') as file:
        for s in updated_students:
            file.write(s + '\n')

    dropdown.options = updated_students

    with output:
        output.clear_output()
        print("Updated successfully")

editButton.on_click(on_edit)
confirmButton.on_click(on_confirm)
```
