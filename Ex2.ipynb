```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import ipywidgets as widgets
import json
from IPython.display import display


students = []

with open('/content/drive/MyDrive/studentsJSON.txt') as f:
  students = json.load(f)

dropdown = widgets.Dropdown(options=[(s["name"], s) for s in students], description="Student:")

confirmButton = widgets.Button(description="Confirm")
editButton = widgets.Button(description="Edit")
idInput = widgets.IntText(description="ID:")
nameInput = widgets.Text(description="Name:")
classInput = widgets.IntText(description="Class:")
emailInput = widgets.Text(description="Email:")

output = widgets.Output()

display(dropdown,confirmButton,idInput,nameInput,classInput,emailInput,editButton,output)

def on_confirm(b):
    selected = dropdown.value

    nameInput.value = selected["name"]
    idInput.value = selected["id"]
    classInput.value = selected["class"]
    emailInput.value = selected["email"]



def on_edit(b):
    updated_students = []

    x = {
     "id": int(idInput.value),
     "name": nameInput.value,
     "class": int(classInput.value),
     "email": emailInput.value,
   }

    for std in students:
        if std["id"] == dropdown.value["id"]:
            updated_students.append(x)
        else:
            updated_students.append(std)

    with open('/content/drive/MyDrive/studentsJSON.txt', 'w') as file:
       json.dump(updated_students,file)

    dropdown.options = [(s["name"], s) for s in updated_students]


    with output:
        output.clear_output()
        print("Updated successfully")

editButton.on_click(on_edit)
confirmButton.on_click(on_confirm)
```
